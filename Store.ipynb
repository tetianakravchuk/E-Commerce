```python
import pandas as pd

# Load the CSV files
revenue = pd.read_csv('revenue.csv')
print(revenue)

targets = pd.read_csv('targets.csv')
print(targets)

men_women = pd.read_csv('men_women_sales.csv')
print(men_women)

print(revenue.columns)
print(targets.columns)
print(men_women.columns)
revenue.rename(columns={'Month': 'month'}, inplace=True)
targets.rename(columns={'Month': 'month'}, inplace=True)
men_women.rename(columns={'Month': 'month'}, inplace=True)

#Strip Extra Spaces in Column Names
revenue.columns = revenue.columns.str.strip()
targets.columns = targets.columns.str.strip()
men_women.columns = men_women.columns.str.strip()

# Merge all the dataframes on the common column 'month'
all_data = revenue.merge(targets, on='month')\
                  .merge(men_women, on='month')

# Print the merged data
print(all_data)


results = all_data[(all_data.revenue > all_data.target) & (all_data.women > all_data.men)]
```

# **Merge on Specific Columns**

```python
import pandas as pd

orders = pd.read_csv('orders.csv')
print(orders)
products = pd.read_csv('products.csv')
print(products)

# Ensure the product_id columns have the same data type
orders['product_id'] = orders['product_id'].astype(int)
products['product_id'] = products['product_id'].astype(int)

orders_products = pd.merge(
    orders,
    products,
    on='product_id'
)

print(orders_products)


```

```python
import pandas as pd

# Example data
orders = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6, 7, 8],
    'product_id': [3, 2, 1, 2, 3, 4, 1, 4],
    'customer_id': [2, 2, 3, 3, 3, 1, 1, 1],
    'quantity': [1, 3, 1, 2, 3, 2, 1, 1],
    'timestamp': [
        '2017-01-01', '2017-01-01', '2017-01-01',
        '2016-02-01', '2017-02-01', '2017-03-01',
        '2017-02-02', '2017-02-02'
    ]
})

products = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'description': ['thing-a-ma-jig', 'whatcha-ma-call-it', 'doo-hickey', 'gizmo'],
    'price': [5, 10, 7, 3]
})

# Merging on product_id from `orders` and id from `products`
orders_products = pd.merge(
    orders,
    products,
    left_on='product_id',  # Match `product_id` from orders
    right_on='id',  # Match `id` from products
    suffixes=['_orders', '_products']  # Add suffixes to overlapping columns
)

# Print the result
print(orders_products)
```

## **Mismatched Merges**

## **Outer Merge**

```python
import pandas as pd

store_a = pd.read_csv('store_a.csv')
print(store_a)
store_b = pd.read_csv('store_b.csv')
print(store_b)

store_a_b_outer = pd.merge(store_a, store_b, how='outer')
print(store_a_b_outer)
```
